**EXPLORATORY DATA ANALYSIS**

```python
numerical_cols = ['overall', 'potential', 'value_eur', 'wage_eur', 'age', 'height_cm', 'weight_kg']
for col in numerical_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col].dropna(), kde=True, bins=30, color="skyblue")
    plt.title(f"Distribution of {col}", fontsize=16)
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

Iterates through numerical columns.

Creates histogram with KDE (density curve) → shows distribution.

Example: overall distribution will be bell-shaped since most players are average.

```python
categorical_cols = ['nationality_name', 'league_name', 'club_name', 'preferred_foot', 'body_type']
for col in categorical_cols:
    plt.figure(figsize=(10,6))
    sns.countplot(data=df, y=col, order=df[col].value_counts().iloc[:10].index, palette="viridis")
    plt.title(f"Top 10 {col} Distribution", fontsize=16)
    plt.xlabel("Frequency")
    plt.ylabel(col)
    plt.show()
```

Loops over categorical columns.

Creates bar charts for top 10 categories (e.g., top 10 nationalities, top clubs).

Useful to see player concentration.

```python
top_players = df.sort_values(by="overall", ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x="overall", y="short_name", data=top_players, palette="magma")
plt.title("Top 10 Highest Overall Players", fontsize=16)
plt.xlabel("Overall")
plt.ylabel("Player Name")
plt.show()
display(top_players[["short_name", "overall","potential","club_name", "nationality_name", "value_eur"]])
```

Sorts players by overall rating.

Selects top 10.

Plots bar chart with player names on y-axis.

Displays their stats (overall, potential, club, nationality, value).

```python
top_players = df.sort_values(by="potential", ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x="overall", y="short_name", data=top_players, palette="magma")
plt.title("Top 10 Highest Overall Players", fontsize=16)
plt.xlabel("Overall")
plt.ylabel("Player Name")
plt.show()
display(top_players[["short_name", "overall","potential", "club_name", "nationality_name", "value_eur"]])
```

Same as before, but sorted by potential instead of overall.

Useful to identify future stars vs current best players.

```python
top_nations = df['nationality_name'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_nations.values, y=top_nations.index, palette="plasma")
plt.title("Top 10 Nationalities by Player Count", fontsize=16)
plt.xlabel("Number of Players")
plt.ylabel("Nationality")
plt.show()
```

Counts number of players by nationality.

Shows top 10 → e.g., Spain, Brazil, Germany usually dominate.

```python
plt.figure(figsize=(12,10))
corr = df[numerical_cols].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix", fontsize=16)
plt.show()
```

Calculates correlation between numerical variables.

Heatmap with annotations (correlation coefficients).

Example insights:

overall highly correlated with potential.

value_eur correlated with overall & potential.

age might have weak or negative correlation with potential.
